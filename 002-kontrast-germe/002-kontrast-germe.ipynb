```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Dosyayı yükle
path = "../images/low_contrast.jpg"
img = cv.imread(path, 0)
```

```python
#histogram hesapla
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])
```

```python
# Histogram grafiğini çiz
plt.plot(histogram)
plt.title('Gri seviyeli (Tek kanallı histogram)')
plt.xlabel('Pixel Değeri')
plt.ylabel('Pixel Sayısı')
plt.show()
```

```python
# histogram eşitleyelim
histogram_esitlenen_foto = cv2.equalizeHist(img)
```

```python
# görseller yan yana getirildi. Farkını görelim
yan_yana_getir = np.hstack((img, histogram_esitlenen_foto))
```

```python
plt.imshow(yan_yana_getir, cmap="gray")
plt.show()
```
